```python
second = df_customer_journey[['interval_of_day','dwell00', 'dwell1','dwell2','dwell3', 'dwell4' ,'dwell5', 'dwell6' ]]
second.head(10)
```

```python
codes = {'morning':1, 'mid-day':2, 'evening':3, 'night':4}
second['interval_of_day'] = second['interval_of_day'].map(codes)
```

```python
second
```

```python
third = second.groupby(['interval_of_day'])[ 'dwell00', 'dwell1','dwell2','dwell3', 'dwell4' ,'dwell5', 'dwell6'          
                                        ].sum().reset_index()
```
